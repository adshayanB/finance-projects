<h1> A Python Script that plots Economic Indicators of the EU 

Importing Packages

```python
import pandas as pd
import requests
import json
import datetime
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

Function to request api and store data

```python
def get_data(link):
    req = requests.get(link)
    req = req.json()
    
    val = req['series']['docs'][0]['value']
    date = req['series']['docs'][0]['period']
    name=req['series']['docs'][0]['dataset_name']
    
    data=pd.DataFrame(val,index=date)
    data.columns=[name]
    
    return data
```

```python
euro_data = get_data('https://api.db.nomics.world/v22/series/Eurostat/irt_euryld_m/M.EA.INS_FWD.CGB_EA.Y10?observations=1')

unemployment_data = get_data('https://api.db.nomics.world/v22/series/Eurostat/une_rt_m/M.NSA.TOTAL.PC_ACT.T.EA19?observations=1')

interest_data = get_data('https://api.db.nomics.world/v22/series/Eurostat/ei_mfir_m/M.NSA.NAP.MF-LTGBY-RT.EU28?observations=1')

inflation_data = get_data('https://api.db.nomics.world/v22/series/WB/WDI/FP.CPI.TOTL.ZG-EU?observations=1')

GDP_data = get_data('https://api.db.nomics.world/v22/series/WB/WDI/NY.GDP.MKTP.KD.ZG-EU?observations=1')

retail_trade_data = get_data('https://api.db.nomics.world/v22/series/Eurostat/sts_trtu_m/M.TOVT.G47.CA.PCH_SM.EA19?observations=1')
# print(retail_trade_data.columns)

```

```python
fig =go.Figure()

fig.add_trace(go.Scatter(
                x=interest_data.index,
                y=interest_data['Interest rates - monthly data'],
                name='INTEREST RATE',
                line_color='Red',
                opacity=0.8))
fig.add_trace(go.Scatter(
                x=euro_data.index,
                y=euro_data['Euro yield curves - monthly data'],
                name='EURO YIELD',
                line_color='deepskyblue',
                opacity=0.8))
fig.add_trace(go.Scatter(
                x=unemployment_data.index,
                y=unemployment_data['Unemployment by sex and age - monthly average'],
                name='UNEMPLOYMENT RATE',
                line_color='green',
                opacity=0.8))
fig.add_trace(go.Scatter(
                x=inflation_data.index,
                y=inflation_data['World Development Indicators'],
                name='INFLATION RATE',
                line_color='orange',
                opacity=0.8))
fig.add_trace(go.Scatter(
                x=GDP_data.index,
                y=GDP_data['World Development Indicators'],
                name='GDP',
                line_color='yellow',
                opacity=0.8))
fig.add_trace(go.Scatter(
                x=retail_trade_data.index,
                y=retail_trade_data['Turnover and volume of sales in wholesale and retail trade - monthly data'],
                name='TRADE DATA',
                line_color='indigo',
                opacity=0.8))





fig.update_layout(xaxis_range=['2001-01-01',datetime.datetime.today()],
                 title_text='Economic Factors',
                  xaxis_rangeslider_visible=True)

fig.update_layout(
    margin=dict(autoexpand=True, l=50, r=50, t=100, b=100, pad=0.5),
    paper_bgcolor="LightSteelBlue"
)
fig.show()
```
