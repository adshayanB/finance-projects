<h1> A Python Script that plots S&P Returns and Stock Price over ten years 

```python
import pandas as pd
import datetime
from datetime import date

import pandas_datareader.data as web
pd.core.common.is_list_like = pd.api.types.is_list_like
import matplotlib.pyplot as plt


```

<h1> Import Data

```python
start_range = datetime.datetime(2010,2,26) 
end_range =datetime.datetime(2020,2,26)

sp500 = web.DataReader(['SP500'],'fred',start_range,end_range)
sp500['Daily Return']=(sp500['SP500']/sp500['SP500'].shift(1))-1 #Calculate Rate of Return 

plt.figure(figsize=(20,10))

sp500['SP500'].plot()
plt.title('S&P 500 Price')
plt.legend()
plt.ylabel('PRICE')

plt.figure(figsize=(20,10))
sp500['Daily Return'].plot()
plt.title('S&P 500 Returns (Daily)')
plt.ylabel('ROR%')

plt.legend()
```
